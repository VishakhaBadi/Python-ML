Check for Missing Values

```python
import pandas as pd


df=pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
df.isnull().sum()
```

a) Drop Rows or Columns

```python
df.dropna(inplace=True)  # drop rows with any missing value
#df.dropna(subset=['Age'], inplace=True)
#df.drop(columns=['Cabin'], inplace=True)
df
```

b) Fill with Static Value

```python
df['Embarked'].fillna('Unknown')
df['Age'].fillna(0)
df
```

c) Fill with Mean, Median, Mode

```python
df['Age'].fillna(df['Age'].mean())
df['Age'].fillna(df['Age'].median())
#df['Embarked'].fillna(df['Embarked'].mode()[0])
df
```

d) Forward/Backward Fill

```python
df.fillna(method='ffill')
df.fillna(method='bfill')
```

**Encoding Categorical Variables**

```python
#df.drop(columns=['cabin'],inplace=True)
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Age']=df['Age'].fillna(df['Age'].mean())
df.info()
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])  # Maleâ†’1, Femaleâ†’0
```

b) One-Hot Encoding

```python
# Label Encoding for binary
df['Sex'] = LabelEncoder().fit_transform(df['Sex'])

# One-Hot Encoding for multi-class
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)
df

```

Final Processing

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
import pandas as pd

# import seaborn as sns
# df = sns.load_dataset('titanic');

df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
df.head()
```

```python
df.info()
```

Drop unused

```python
df.drop(columns=['Cabin'], inplace=True)
df.head()
```

Fill missing

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df.info()
```

Encode

```python
le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)
df.head()
```

Scale

```python
scaler = StandardScaler()
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])
df
```

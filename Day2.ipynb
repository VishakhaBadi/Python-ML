Create Arrays

```python
import numpy as np

a=np.array([1,2,3,4,5,6])
b=np.arange(1,6)
c=np.zeros((3,3))
d=np.ones((4,4))

print("array is:",a)
print("range is:",b)
print("zeros is:",c)
print("ones is:",d)
```

Indexing & Slicing

```python
print("indexing is:",a[3])
print("silicing is:",a[-2:])
```

Basic arithmetic operations

```python
print("it adds to all numbers:",a+2)
print("it multifes to all:",a*2)
```

Basic statistics

```python
print("square of a:",np.square(a))
print("means of a:",np.mean(a))
print("standerd of a:",np.std(a))
```

Reshaping arrays

```python
m=np.arange(1,10).reshape((3,3))
print(m)
```

Transpose

```python
print(m.T)
```

Boolean indexing

```python
print(m[m>5])
print(m>5)
```

```python
import pandas as pd
data={
    'name':['vani','nayii','eshii','ranii','billu'],
    'age':[20,19,13,24,16],
    'department':['bca','ba','bcom','bca','be'],
    'salary':[20000,34000,45640,35666,10000]

}
df=pd.DataFrame(data)
df
```

Basic operations

```python
print(" first 3 rows")
df.tail(2)
```

```python
print("columns")
df.columns
```

```python
print("data types")
df.dtypes
```

```python
print("discribe")
df.describe()
```

Filtering

```python
print("filtering")
df[df['salary']>40000]
```

Sorting

```python
print("sorting ")
df.sort_values(by='age')
```

Grouping and Aggregation

```python
group=df.groupby('department')['salary'].sum()
print("sum of each dept ")
group
```

Adding new columns

```python
print("adding new culumns")
df['avg of salary']=df['salary']/2
df
```

**üìä Matplotlib: Building Blocks of Visualizations**

**Common Charts:**
Line Plot
Bar Chart
Histogram
Pie Chart

line chart

```python
import matplotlib.pyplot as plt
years=[2020,2021,2022,2026,2030]
revenue=[000,400,150,300,600]

plt.plot(years,revenue,marker='v',linestyle='--',color='yellow')
plt.title("years of revenue")
plt.xlabel("years")
plt.ylabel("revenue ")
plt.show()
```

Bar Chart

```python
import matplotlib.pyplot as plt

department=df['department'].value_counts()
department.plot(kind='bar' ,color='pink')
plt.title("bar chart")
plt.xlabel("departments")
plt.ylabel("counts")
plt.show()
```

Histogram

```python
import matplotlib.pyplot as plt
ages=df['age']
plt.hist(ages,bins=5, color='coral', edgecolor='black')
plt.title("historigram")
plt.xlabel("frequency")
plt.show()
```

Pie Chart

```python
df.groupby('department').size().plot(kind='pie', autopct='%1.1f%%')
plt.title("Department Proportion")
plt.ylabel("")
plt.show()
```

**üåà Seaborn: Statistical Data Visualization**


Load sample dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt


exercise = sns.load_dataset("exercise")
exercise.head()
```

```python
exercise.info()
```

```python
exercise.describe(include='all')
```

Pairplot

```python
sns.pairplot(exercise, hue="kind")
plt.show()
```

Heatmap of correlation

```python
corr = exercise.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

Countplot

```python
sns.countplot(x='kind', data=exercise, palette='Set2', hue='kind', legend=False)
plt.title("Count of Records by Day")
plt.show()
```

Boxplot

```python
sns.boxplot(x='kind', y='pulse', data=exercise)
plt.title("Boxplot of Total Bill by Day")
plt.show()
```
